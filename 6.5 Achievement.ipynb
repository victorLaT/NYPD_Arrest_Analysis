# 6.5 K Means Unsupervised assignment

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import scipy
import folium
import json
import sklearn
from sklearn.cluster import KMeans
import pylab as pl
```

```python
path = r'Desktop/Crimes_by_Prec.pkl'
```

```python
df = pd.read_pickle(os.path.join(path))
```

```python
df.head()
```

```python
df.head()
```

```python
%matplotlib inline
```

```python
num_cl = range(1, 10) 
kmeans = [KMeans(n_clusters=i) for i in num_cl] 
```

```python
score = [kmeans[i].fit(df).score(df) for i in range(len(kmeans))]
```

### With this elbow curve I was able to see that there are about 4 clusters

```python
pl.plot(num_cl,score)
pl.xlabel('Number of Clusters')
pl.ylabel('Score')
pl.title('Elbow Curve')
pl.show()
```

```python
df.head()
```

```python
kmeans = KMeans(n_clusters = 3)
```

```python
kmeans.fit(df)

df['clusters'] = kmeans.fit_predict(df)

df['clusters'].value_counts()
```

```python
## This cluster plot tells me that the usual occurences of crime are dispersed while there are many more occurences in the precincts under 60.
```

```python
plt.figure(figsize=(12,8))
ax = sns.scatterplot(x=df['DANGEROUS DRUGS'], y=df['ARREST_PRECINCT'], hue=kmeans.labels_, s=100) 

ax.grid(False) 
plt.xlabel('DANGEROUS DRUGS') 
plt.ylabel('ARREST_PRECINCT') 
plt.show()
```

```python
df.loc[df['clusters'] == 2, 'cluster'] = 'dark purple'
df.loc[df['clusters'] == 1, 'cluster'] = 'purple'
df.loc[df['clusters'] == 0, 'cluster'] = 'pink'

df.groupby('cluster').agg({'ASSAULT 3 & RELATED OFFENSES':['mean', 'median'], 
                         'DANGEROUS DRUGS':['mean', 'median'], 
                         'GRAND LARCENY OF MOTOR VEHICLE':['mean', 'median'],
                          'VEHICLE AND TRAFFIC LAWS':['mean', 'median']})
```

## Judging by these statistics the purple cluster seem to have the most impact on Crime. The median is higher and
